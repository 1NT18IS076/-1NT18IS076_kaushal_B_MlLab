```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('../data/UCL_zoo_dataset.csv')
```

```python
df.head()
```

```python
classes = ['Mammal', 'Bird', 'Reptile', 'Fish', 'Amphibian', 'Bug', 'Invertebrate']
```

```python
df.describe()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df.drop('class_type', axis=1).drop('animal_name', axis=1), 
    df['class_type'], 
    test_size=0.33, random_state=42
)
```

```python
len(X_train)
```

```python
X_train.head()
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()

model.fit(X_train, y_train)

tree.plot_tree(model)
```

```python
X_test.head()
```

```python
pred = model.predict(X_test)
```

```python
pred[:10]
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
predprint(confusion_matrix(y_test, pred))
```

```python
print(classification_report(y_test, pred, target_names=classes))
```

```python
print("Actual\tPrediction")
y_test = list(y_test)

for i in range(len(y_test[:25])):
    print(f"{classes[y_test[i]-1]}\t{classes[pred[i]-1]}")
```
